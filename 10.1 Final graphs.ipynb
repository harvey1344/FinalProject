```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
final_df= pd.read_csv('algo_data/final1m')
final_df.drop(inplace=True, columns=['Unnamed: 0'])
final_df
```

```python

plt.scatter(final_df['Fit Time'], final_df['RMSE'])
for i, row in final_df.iterrows():
    plt.text(row['Fit Time'], row['RMSE'], row['Algorithm'])

plt.xlabel('Fit Time')
plt.ylabel('RMSE')
plt.title('Fit Time vs. RMSE')
plt.savefig('./final_fit.png')
plt.show()
```

```python
plt.scatter(final_df['Predict Time'], final_df['RMSE'])
for i, row in final_df.iterrows():
    plt.text(row['Predict Time'], row['RMSE'], row['Algorithm'])

plt.xlabel('Predict time')
plt.ylabel('RMSE')
plt.title('Fit Time vs. RMSE')

plt.savefig('./final_predict.png')
plt.show()
```
